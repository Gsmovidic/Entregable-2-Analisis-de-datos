```python
import pandas as pd
```

# Cargar datos

```python
import kagglehub

# Download latest version
path = kagglehub.dataset_download("arianazmoudeh/airbnbopendata")

print("Path to dataset files:", path)
```

```python
df = pd.read_csv(
    path + '/Airbnb_Open_Data.csv',
    low_memory=False)
df.head()
```

```python
df.columns
```

```python
df.info()
```

```python
df['neighbourhood group'].unique()
```

```python
df['neighbourhood group'].value_counts()
```

# Limpieza de datos

# Corregir datos: brookln, manhatan

```python
df['neighbourhood group'] = df['neighbourhood group'].replace({
    'brookln': 'Brooklyn',
    'manhatan': 'Manhattan'
})
```

```python
df['neighbourhood group'].value_counts()
```

# Normalizar nombres columnas para que cuando haya espacio este un _

```python
df.columns = df.columns.str.strip().str.lower().str.replace(' ', '_')
```

#  Limpiar price y service_fee

```python
df['price'] = df['price'].str.replace('$', '', regex=False).str.replace(',', '', regex=False).astype(float)
df['service_fee'] = df['service_fee'].str.replace('$', '', regex=False).str.replace(',', '', regex=False).astype(float)
```

# Convertir fechas

```python
df['last_review'] = pd.to_datetime(df['last_review'])
```

# Convertir booleanos

```python
df['instant_bookable'] = df['instant_bookable'].map({'TRUE': 1, 'FALSE': 0})
```

# Rellenamos valores vacios

```python
df['reviews_per_month'] = df['reviews_per_month'].fillna(0)
```

# Eliminar columnas que no nos sirven

```python
df = df.drop(columns=['license', 'house_rules','country'])
```

# Convertir columnas numericas

```python
cols = [
    'minimum_nights',
    'number_of_reviews',
    'reviews_per_month',
    'review_rate_number',
    'calculated_host_listings_count',
    'availability_365'
]

df[cols] = df[cols].apply(pd.to_numeric, errors='coerce')
```

```python
df.info()
```

```python
df['room_type'].value_counts()
```

# Guardar dataframe limpio

```python
df.to_csv('airbnb_limpio.csv', index=False)
```
